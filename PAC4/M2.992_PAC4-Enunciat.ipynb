```python
# -- DD --
num_graphs_dd = dd_graph_labels.shape[0]
num_classes_dd = dd_graph_labels["Class_label"].nunique()
# - Max nodes i mitja -
ids_dd = dd_graph_indicator["Graph_id"].tolist() # Bibliografia: [Ref.4]
dict_graph_nodes_dd = {} # {graph_id: count_nodes}
for id_val in ids_dd:
    if id_val in dict_graph_nodes_dd:
        dict_graph_nodes_dd[id_val] += 1
    else:
        dict_graph_nodes_dd[id_val] = 1
max_nodes_dd = max(dict_graph_nodes_dd.values())
average_nodes_dd = sum(dict_graph_nodes_dd.values()) / len(dict_graph_nodes_dd)
average_edges_dd = (len(dd_a) / num_graphs_dd) / 2 # Dividir per 2 per cada aresta apareix 2 cops
```

```python
# -- ENZYMES --
num_graphs_enzymes = enzymes_graph_labels.shape[0]
num_classes_enzymes = enzymes_graph_labels["Class_label"].nunique()
# - Max nodes i mitja -
ids_enzymes = enzymes_graph_indicator["Graph_id"].tolist() # Bibliografia: [Ref.4]
dict_graph_nodes_enzymes = {} # {graph_id: count_nodes}
for id_val in ids_enzymes:
    if id_val in dict_graph_nodes_enzymes:
        dict_graph_nodes_enzymes[id_val] += 1
    else:
        dict_graph_nodes_enzymes[id_val] = 1
max_nodes_enzymes = max(dict_graph_nodes_enzymes.values())
average_nodes_enzymes = sum(dict_graph_nodes_enzymes.values()) / len(dict_graph_nodes_enzymes)
average_edges_enzymes = (len(enzymes_a) / num_graphs_enzymes) / 2 # Dividir per 2 per cada aresta apareix 2 cops
```

```python
# -- PROTEINS --
num_graphs_proteins = proteins_graph_labels.shape[0]
num_classes_proteins = proteins_graph_labels["Class_label"].nunique()
# - Max nodes i mitja -
ids_proteins = proteins_graph_indicator["Graph_id"].tolist() # Bibliografia: [Ref.4]
dict_graph_nodes_proteins = {} # {graph_id: count_nodes}
for id_val in ids_proteins:
    if id_val in dict_graph_nodes_proteins:
        dict_graph_nodes_proteins[id_val] += 1
    else:
        dict_graph_nodes_proteins[id_val] = 1
max_nodes_proteins = max(dict_graph_nodes_proteins.values())
average_nodes_proteins = sum(dict_graph_nodes_proteins.values()) / len(dict_graph_nodes_proteins)
average_edges_proteins = (len(proteins_a) / num_graphs_proteins) / 2 # Dividir per 2 per cada aresta apareix 2 cops
```
