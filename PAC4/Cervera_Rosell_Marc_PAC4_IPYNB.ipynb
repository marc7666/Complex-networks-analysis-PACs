````python
graphs = {graph_id: nx.Graph() for graph_id in range(1, num_graphs_proteins + 1)} # +1 perque comencem a crear a aprtir del graf 1 i no del graf 0
graphs_nodes = {} # Bibliografia: [Ref.6] -> Ús 2
# - Afegir nodes a cad graph -
for index, graph_id in proteins_graph_indicator.iterrows(): # Bibliografia: [Ref.7]
    graph = graphs[graph_id.iloc[0]] # Apunt 1
    graph.add_node(index + 1)
    if index + 1 not in graphs_nodes: # Bibliografia: [Ref.6] -> Ús 2
        graphs_nodes[index + 1] = graph_id.iloc[0]
    
# - Afegir arestes -
# Bibliografia: [Ref.6] -> Ús 2
for index, nodes in proteins_a.iterrows():
    node1, node2 = nodes.iloc[0], nodes.iloc[1]
    graph_to_add_edge = graphs_nodes[node1]
    actual_graph = graphs[graph_to_add_edge]
    if not actual_graph.has_edge(node1, node2):
        actual_graph.add_edge(node1, node2)
````

````python
# -- Grau mitjà --
average_degree_dict = {} # {graph_id: average_degree}
for graph_id, graph in graphs.items():
    graph_nodes = graph.number_of_nodes()
    average_degree = sum(dict(graph.degree()).values()) / graph_nodes # Bibliografia: [Ref.8]
    average_degree_dict[graph_id] = average_degree
````

````python
# -- longitud mitjana del camí més curt --
avg_shortest_path_length_dict = {} # {graph_id: average_shortest_path_lenght de la component connexa mes gran}
for graph_id, graph in graphs.items():
    # Es calcula la component connexa mes gran com a sistema de seguretat per si el graf sencer no fos connex.
    biggest_connected_component = max(nx.connected_components(graph), key=len) # Bibliografia: [Ref.8]
    subgraph_biggest_connected_component = graph.subgraph(biggest_connected_component) # Bibliografia: [Ref.9]
    avg_shortest_path_length_dict[graph_id] = nx.average_shortest_path_length(subgraph_biggest_connected_component) # Bibliografia: [Ref.10]
````

````python
model = GATProteins()
optimizer = torch.optim.Adam(model.parameters(), lr=0.0005, weight_decay=0.001) # Bibliografia: [Ref.19], [Ref.24] i [Ref.6] -> Ús 7
loss_function = torch.nn.CrossEntropyLoss() # Bibliografia: [Ref.19], [Ref.25] i [Ref.26]
````

<div style="background-color: #d2ffc9; border-color: #004000; border-left: 5px solid #004000; padding: 0.5em;">

````python
# Bibliografia: [Ref.6] -> Ús 8
training_place = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = model.to(training_place)
# Bibliografia: [Ref.27], [Ref.28] i [Ref.6] -> Ús 9
def train():
    model.train()
    total_loss = total_examples = 0
    for data in train_dataloader:
        data = data.to(training_place)
        data.x = data.x.float()
        optimizer.zero_grad()
        out = model(data)
        loss = loss_function(out, data.y)
        loss.backward()
        optimizer.step()
        batch = data.y.size(0)
        total_loss += loss.item() * batch
        total_examples += batch
    return total_loss / total_examples
````

````python
# Bibliografia: [Ref.28]
def validation(loader):
    model.eval()
    correct = 0
    for data in loader:
        data = data.to(training_place) # Bibliografia: [Ref.6] -> Ús 8
        data.x = data.x.float()
        out = model(data)
        pred = out.argmax(dim=1)
        correct += int((pred == data.y).sum())
    return correct / len(loader.dataset)
````

````python
# Bibliografia: [Ref.30]
from sklearn.model_selection import KFold # Per fer els folds # Bibliografia: [Ref.31]

kfold = KFold(n_splits=5, shuffle=True, random_state=42)
accuracies = []
epochs = 300

for fold, (train_ids, test_ids) in enumerate(kfold.split(dataset_torch_geom)):
    print(f"Fold {fold}")
    # -- DataLoader per la fold --
    # - Longituds -
    train_len = int(len(dataset_torch_geom) * 0.8)
    validation_len = int(len(dataset_torch_geom) * 0.1)
    test_len = len(dataset_torch_geom) - train_len - validation_len
    # - Split - # Bibliografia: [Ref.6] -> Ús 11
    test_set = [dataset_torch_geom[i] for i in test_ids]
    train_validation_set = [dataset_torch_geom[i] for i in train_ids]
    validation_set_len = (len(train_validation_set) * 10) // 100 # 10% per al set de validació
    train_set = train_validation_set[validation_set_len:]
    validation_set = train_validation_set[:validation_set_len]     
    # - DataLoaders -
    train_dataloader = DataLoader(train_set, batch_size=32, shuffle=True)
    validation_dataloader = DataLoader(validation_set, batch_size=32, shuffle=False)
    test_dataloader = DataLoader(test_set, batch_size=32, shuffle=False)
    # -- Definir el model i l'optimitzador ---
    current_model = GATProteins().to(training_place)
    optimizer = torch.optim.Adam(current_model.parameters(), lr=0.0009, weight_decay=0.001)
    # -- Entrenar i guardar el millor --
    best_checkpoint = 0
    for epoch in tqdm.auto.trange(1, epochs + 1):
        train2(current_model, train_dataloader, optimizer, epoch)
        validation_acc = validation2(current_model, validation_dataloader)
        if validation_acc > best_checkpoint:
            best_checkpoint = validation_acc
            best_model_name = f"best_model_k_fold_gat_fold{fold}.pth"
            torch.save(current_model.state_dict(), best_model_name)
    # -- Validar amb el millor --
    current_model.load_state_dict(torch.load(best_model_name))
    test_accuracy = validation2(current_model, test_dataloader)
    accuracies.append(test_accuracy)
````
