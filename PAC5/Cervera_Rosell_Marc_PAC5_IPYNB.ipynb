````python
import pandas as pd
import networkx as nx
from networkx.algorithms.community import louvain_partitions
from collections import Counter
````

<div style="background-color: #d2ffc9; border-color: #004000; border-left: 5px solid #004000; padding: 0.5em;">
<strong>Lectura de fitxers</strong>
</div>

````python
# -- Lectura del fitxer edges.csv --
edges = pd.read_csv("edges.csv", sep=",", header=None) # Bibliografia: [Ref.1]
edges.rename(columns={0: "Node_source", 1: "Node_target"}, inplace=True) # Bibliografia: [Ref.2]
edges
````

````python
# -- Lectura del fitxer node_classification_train.csv --
node_classification_train = pd.read_csv("node_classification_train.csv", sep=",", header=None) # Bibliografia: [Ref.1]
node_classification_train.rename(columns={0: "Node_name", 1: "Node_classification"}, inplace=True) # Bibliografia: [Ref.2]
node_classification_train
````

````python
# -- Lectura del fitxer node_features.csv --
node_features = pd.read_csv("node_features.csv", sep=",", header=None) # Bibliografia: [Ref.1]
for i in range(node_features.shape[1]):
    node_features.rename(columns={i: f"Feature_{i}"}, inplace=True) # Bibliografia: [Ref.2]
node_features
````

````python
# -- Lectura del fitxer node_test.csv --
node_test = pd.read_csv("node_test.csv", sep=",", header=None) # Bibliografia: [Ref.1]
node_test.rename(columns={0: "Node"}, inplace=True) # Bibliografia: [Ref.2]
node_test
````

<div style="background-color: #d2ffc9; border-color: #004000; border-left: 5px solid #004000; padding: 0.5em;">
<strong>Construcció del graf</strong>
</div>

````python
edges_list = []
for index, row in edges.iterrows(): # Bibliografia: [Ref.3]
    edge_tuple = (int(row.iloc[0]), int(row.iloc[1])) # Apunt 1
    edges_list.append(edge_tuple)
````

**Apunt 1:** Si no s'utilitza el _iloc_ salta un warning de futur amb suggeriment de correcció.

És a dir, si es crea la tupla com:

```python

(row["Node_source"], row["Node_target"])

```

Salta el warning

````python
graph = nx.Graph()
graph.add_edges_from(edges_list) # Bibliografia: [Ref.4]
````

````python
print(f"Arestes del graf: {graph.number_of_edges()}") # Bibliografia: [Ref.5]
print(f"Nodes del graf: {graph.number_of_nodes()}") # Bibliografia: [Ref.6]
````

<div style="background-color: #fcf2f2; border-color: #dfb5b4; border-left: 5px solid #dfb5b4; padding: 0.5em;">
  
El nombre d'arestes es correspon amb el nombre de línies de edges.csv.

El nombre de nodes es correspon amb el nombre de línies de node_features.csv

Atès que el nombre de arestes i el nombre de nodes es correspon amb els fitxers proporcionats a les dades, es conclou que el graf ha estat creat correctament.

</div>

````python
# -- Afegir features als nodes--
mapping_nodes_features = node_features.to_dict(orient="index") # Bibliografia: [Ref.7]
nx.set_node_attributes(graph, mapping_nodes_features, "features") # Bibliografia: [Ref.8] i [Ref.9]
````

````python
# -- Afegir features les classificacions conegudes --
mapping_node_classification = {} #to_dict -> No útil aquí - Cap 'orient' és útil
for index, row in node_classification_train.iterrows():
    node = int(row.iloc[0])
    classification = int(row.iloc[1])
    mapping_node_classification[node] = classification
nx.set_node_attributes(graph, mapping_node_classification, "classification")
````

<div style="background-color: #d2ffc9; border-color: #004000; border-left: 5px solid #004000; padding: 0.5em;">
<strong>Predicció d'etiquetes mitjançant detecció de comunitats i propagació d'etiquetes</strong>
</div>

<div style="background-color: #fcf2f2; border-color: #dfb5b4; border-left: 5px solid #dfb5b4; padding: 0.5em;">
  
Per resoldre aquesta pràctica s'han reutilitzat els coneixements de la PAC2 sobre detecció de comunitats i propagació d'etiquetes.

S'ha utilitzat la tècnica de detecció de comunitats i propagació d'etiquetes sota la premisa de que investigadors relacionats (que interactuen) tendeixen a treballar en les mateixes àrees.

S'ha pogut comprovar que després del pas "Assignació de classificacions als nodes per "vot popular" (classificació majoritària de la comunitat)" encara quedaven alguns nodes sense classificar. A aquests nodes sel's ha assignat la classificació majoritària del graf (pas "Assignació de classificacions als nodes que queden sense etiqueta (s'assigna la classificació majoritària)")

<strong>Detecció de comunitats mitjançant l'algorisme de detecció de comunitats de Louvain</strong>

<div style="background-color: #fcf2f2; border-color: #dfb5b4; border-left: 5px solid #dfb5b4; padding: 0.5em;">

````python
louvain = list(louvain_partitions(graph, seed=42))  # Bibliografia: [Ref.10] i [Ref.11] -> Ús 1
communities = []
modularities = []
for elem in louvain: # Càlcul de la mida de les comunitats i les modularitats
    num_comms = len(elem)
    modularity = nx.community.modularity(graph, elem)
    communities.append(num_comms)
    modularities.append(modularity)
````

````python
communities
````

````python
modularities
````

<div style="background-color: #9ECEDD; border-color: #419EBB; border-left: 5px solid #419EBB; padding: 0.5em;">
<strong>Tria de la millor partició i assignació de comunitat als nodes</strong>
</div>

````python
# Tria de la partició amb més modularitat. A Bibliografia: [Ref.10] explica que com més modularitat, millor es la partició
best_partition = louvain[modularities.index(max(modularities))] # Bibliografia: [Ref.12] i [Ref.11] -> Ús 2
mapping_node_comm = {} # {node : comm}
for comm_num, nodes_comm in enumerate(best_partition):
    for node in nodes_comm:
        mapping_node_comm[node] = comm_num

nx.set_node_attributes(graph, mapping_node_comm, "community") # Assignar comunitat als nodes
````

<div style="background-color: #9ECEDD; border-color: #419EBB; border-left: 5px solid #419EBB; padding: 0.5em;">
<strong>Mapping de les classificacions de cada comunitat</strong>
</div>

````python
mapping_community_classifications = {} #{community_num:[classifications]}
nodes = graph.nodes()
for node in nodes:
    if "classification" in nodes[node]:
        community = nodes[node]["community"]
        if community not in mapping_community_classifications.keys():
            mapping_community_classifications[community] = [] # Per guardar les classificacions
        mapping_community_classifications[community].append(nodes[node]["classification"])
````

<div style="background-color: #9ECEDD; border-color: #419EBB; border-left: 5px solid #419EBB; padding: 0.5em;">
<strong>Mapping de les classificacions majoritàries de cada comunitat</strong>
</div>

````python
most_repeated_label_community = {} # {community_num: most_repeated_label}
for community, labels in mapping_community_classifications.items():
    count = Counter(labels) # Bibliografia: [Ref.13]
    most_repeated_label_community[community] = count.most_common(1)[0][0] # Etiqueta més repetida # Bibliografia: [Ref.14]
````

<div style="background-color: #9ECEDD; border-color: #419EBB; border-left: 5px solid #419EBB; padding: 0.5em;">
<strong>Assignació de classificacions als nodes per "vot popular" (classificació majoritària de la comunitat)</strong>
</div>

````python
for node in nodes:
    if "classification" not in nodes[node]:
        community = nodes[node]["community"]
        if community in most_repeated_label_community.keys():
            nodes[node]["classification"] = most_repeated_label_community[community]
````

<div style="background-color: #9ECEDD; border-color: #419EBB; border-left: 5px solid #419EBB; padding: 0.5em;">
<strong>Assignació de classificacions als nodes que queden sense etiqueta (s'assigna la classificació majoritària) --> Idea extreta de # Bibliografia: [Ref.11] -> Ús 3</strong>
</div>

````python
training_classifications = node_classification_train["Node_classification"].tolist() # Bibliografia: [Ref.15]
count_classifications_graph = Counter(training_classifications)
most_repeated_label_graph = count_classifications_graph.most_common(1)[0][0] # Etiqueta més repetida

# -- Assignació de classificació als nodes no classificats --
for node in nodes:
    if "classification" not in nodes[node]:
        nodes[node]["classification"] = most_repeated_label_graph
````

<div style="background-color: #9ECEDD; border-color: #419EBB; border-left: 5px solid #419EBB; padding: 0.5em;">
<strong>Comprovació de que no queden nodes sense classificar</strong>
</div>

````python
no_classified_nodes = []
for node in nodes:
    if "classification" not in nodes[node]:
        no_classified_nodes.append(node)
````

<div style="background-color: #9ECEDD; border-color: #419EBB; border-left: 5px solid #419EBB; padding: 0.5em;">
<strong>Guardar resultats en fitxer per a Kaggle</strong>
</div>

````python
# Bibliografia: [Ref.16]
results_kaggle = []
for index, row in node_test.iterrows():
    node = row.iloc[0]
    classification = nodes[node]["classification"]
    results_kaggle.append((node, classification))

results_kaggle_df = pd.DataFrame(results_kaggle, columns=["id", "label"])
results_kaggle_df.to_csv("results.csv", index=False) # Bibliografia: [Ref.17]
````
