```python
#Construcció del graf base amb la funció watts_strogatz_graph() de la llibreria networkx amb 20 nodes, grau4 i probailitat 0
N = 20 # Nodes del graf base
K = 4 # Grau dels nodes
b = 0 # Prbabilitat beta
base = nx.watts_strogatz_graph(N, K, b) # Bibliografia: [1]
plt.figure()
plt.title("Model  beta amb 20 nodes, grau 4 i probabilitat beta 0")
nx.draw(base, with_labels=True) # Bibliografia: [2] i [7]
plt.show()
```

```python
# Procediment per recablejar el graf
import random as rnd # Generacio nombres aleatoris --> generacio de nombres entre 0.0 i 1.0 per saber si recablejar i per eleccio random de nou node
def rewiring_beta_model(graph, p):
    x = graph.copy() # Bibliografia: [3]
    nodes = x.nodes() # Tots els nodes del graf base # Bibliografia: [4]
    edges = x.edges() # Totes les arestes del graf base # Bibliografia: [5]
    #p = 0.7 # Probabilitat de recablejar
    """
    Passos rewiring:
    1.- Trobar els K/2 veins a la dreta.
    2.- Si hi ha aresta entre el node que s'esta tractant i el veí i la probabilitat (generada random) < p (probabilitat de rewire) -> rewire
        2.1.- El nou node no pot ser ell mateix
        2.2.- El nou node no pot ser un veí que ja te.
    """
    for node in nodes:
        # Step 1 - Buscar els K/2 veins de la dreta
        rightmost_neighbours = [] # Per guardar els veins
        for vei in range(1, K // 2 + 1): # range(1, 4 // 2 + 1) = [1, 2]
            rightmost_neighbours.append((node + vei) % N) # Calcul dels veins de la dreta
            # Step 2 - Si hi ha aresta entre el node que s'esta tractant i els veins i la probabilitat random < probabilitat de rewire
            for neighbour in rightmost_neighbours:
                if x.has_edge(node, neighbour) and round(rnd.uniform(0.0, 1.0), 1) < p: # Bibliografia: [8] [9] [10] [11]
                    new_possible_connections = nodes - x.neighbors(node) - {node} # Els nous possibles nodes son tots - ell mateix - els veins actuals
                    x.remove_edge(node, neighbour) # Descablejar el node i el vei # Bibliografia: [12]
                    new_node = rnd.choice(list(new_possible_connections)) # Bibliografia: [13]
                    x.add_edge(node, new_node) # Bibliografia [14]
        return x
```

```python
rewired = rewiring_beta_model(base, 0.7)
plt.figure(figsize=(15,15)) # Bibliografia: [16]
plt.title("Model beta recablejat")
nx.draw(rewired, with_labels=True)
plt.tight_layout() # Bibliografia: [17]
plt.show()
```

```python
# Recreació de la imatge 3.7 del llibre
N = 200
K = 4
times_rep = 100 # Nombre de repeticions per cada beta
beta = [round(i, 2) for i in np.arange(0.0, 1.5, 0.05)] # Probabilitats beta de 0.0 a 1.0 amb un salt de 0.05 # Bibliografia: [18]
path_length = [] # Llista per guardar les L a graficar no normalitzades --> Mitjana de cada beta
clustering_coeff = [] # Llista per guardar les C a graficar no normalitzades --> Mitjana de cada beta
for prob in beta:
    paths_every_beta = []
    clustering_coeff_every_beta = []
    for i in range(times_rep):
        graph = nx.watts_strogatz_graph(N, K, prob) # Generacio del graf base
        paths_every_beta.append(nx.average_shortest_path_length(graph)) # Per cada repeticio ens quedem amb la longitud mitjana del cami mes curt
        # Bibliografia: [19]
        clustering_coeff_every_beta.append(nx.average_clustering(graph)) # Per cada repeticio ens quedem amb el coeficient mitja del graf
        # Bibliografia: [20]
    path_length.append(np.mean(paths_every_beta)) # Bibliografia: [21]
    clustering_coeff.append(np.mean(clustering_coeff_every_beta))
path_length_norm= [] # L normalitzades
# Bibliografia: [22]
for i in range(len(path_length)):
    path_length_norm.append((path_length[i] - min(path_length)) / (max(path_length) - min(path_length)))
clustering_coeff_norm = [] # C normalitzades
for j in range(len(clustering_coeff)):
    clustering_coeff_norm.append((clustering_coeff[j] - min(clustering_coeff)) / (max(clustering_coeff) - min(clustering_coeff)))
```

```python
plt.figure()
plt.plot(beta, path_length_norm, color = "red", label="L") # Bibliografia: [15] i [24]
plt.plot(beta, clustering_coeff_norm, color = "green", label="C") # Bibliografia: [15] i [24]
plt.xlabel("Beta")  # Bibliografia: [23]
plt.ylabel("Path length")  # Bibliografia: [23]
plt.title("Path length and clustering coefficient in the beta model")  # Bibliografia: [23]
plt.legend() # Bibliografia: [25]
plt.show()
```

```python
import random as rnd
# Bibliografia: [27]
# m = Nombre de noves connexions
# final_graph_nodes = Nombre de nodes amb finals --> Condicio per parar l'algorisme
def barabasi_albert_preferential_attachement(graph, m, final_graph_nodes):
    final_graph = graph.copy() # Bibliografia: [3]
    initial_nodes = len(list(graph.nodes()))
    for i in range(initial_nodes, final_graph_nodes):
        node_degrees = dict(final_graph.degree()) # Diciconari {node: grau} # Bibliografia: [28]
        connections_of_new_node = rnd.choices(list(node_degrees.keys()), weights=list(node_degrees.values()), k=m) # Bibliografia: [29] i [30]
        # weights = possibilitat per cada node. k = llarg de la llista reotrn
        final_graph.add_node(len(list(nx.nodes(final_graph)))) # len(list(nx.graph(nodes))) - Nom node = següent nombre # Bibliografia: [31] i [32]
        for node in connections_of_new_node:
            final_graph.add_edge(node, i)
    return final_graph
```

```python
# Construcció imatge 5.4
random_graph = nx.Graph()
random_graph.add_nodes_from(range(5))# Bibliografia: [33]
while not nx.is_connected(random_graph): # Bibliografia: [34]
    node1 = rnd.randint(0, 4) # Bibliografia: [35]
    node2 = rnd.randint(0, 4) # Bibliografia: [35]
    random_graph.add_edge(node1, node2)

m = 5
final_nodes = 10000
final_graph = barabasi_albert_preferential_attachement(random_graph, m, final_nodes)
# --- Trobar probabilitats de cada grau i grafic ----- # Bibliografia: [37]
degrees_mapping = final_graph.degree() # Mapping node: grau # Bibliografia: [38]
degrees_list = list(dict(degrees_mapping).values()) # Llista de graus --> Casos possibles
degrees_cardinality = {} # Mapping grau d'un node: nombre de vegades que existeix el grau en el graf (casos favorables)
for num in degrees_list:
    if num in degrees_cardinality.keys():
        degrees_cardinality[num] += 1
    else:
        degrees_cardinality[num] = 1
possible_cases = len(degrees_list) # Total de casos possibles
probabilities = {} # grau d'un node: probabilitat
for i in degrees_cardinality.keys():
    probabilities[i] = degrees_cardinality[i] / possible_cases

# ----- Log-Binning ----- # Bibliografia: [41] i [42]
max_x = np.log10(max(probabilities.keys())) # Bibliografia: [43]
max_y = np.log10(max(probabilities.values())) # Bibliografia: [43]
max_base = max([max_x, max_y])
min_x = np.log10(min([key for key in probabilities.keys() if key!=0])) # Bibliografia: [43]
bins = np.logspace(min_x, max_base, num=35) # Bibliografia: [44]
keys = np.array(list(probabilities.keys())) # Bibliografia: [45]
vals = np.array(list(probabilities.values())) # Bibliografia: [45]
bin_means_x = (np.histogram(keys, bins, weights=keys)[0] / np.histogram(keys, bins)[0]) # Bibliografia: [46] i [15]
bin_means_y = (np.histogram(keys, bins, weights=vals)[0] / np.histogram(keys, bins)[0]) # Bibliografia: [46] i [15]

plt.figure(figsize=(15,10))
plt.loglog(probabilities.keys(), probabilities.values(), 'o') # Bibliografia: [37] i [39]
plt.loglog(bin_means_x, bin_means_y, 'x', color='red') # Bibliografia: [47]
plt.ylabel("Pk", fontsize=18) # Bibliografia: [40]
plt.xlabel("k", fontsize=18) # Bibliografia: [40]
plt.show()
```

```python
edges_power_grid = []
with open("powergrid.edgelist.txt") as f:
    for line in f:
        splitted_data = line.strip().split("\t") # Bibliografia: [48] [49]
        edges_power_grid.append((splitted_data[0], splitted_data[1]))
    f.close()
edges_power_grid = edges_power_grid[1:]
```

```python
nodes_power_grid = []
for pairs in edges_power_grid:
    if pairs[0] not in nodes_power_grid:
        nodes_power_grid.append(pairs[0])
    elif pairs[1] not in nodes_power_grid:
        nodes_power_grid.append(pairs[1])
```

```python
# ----- Generar graf power grid -----
power_grid = nx.Graph()
# ----- Afegir nodes -----
power_grid.add_nodes_from(nodes_power_grid)
# ----- Afegir arestes -----
for pair_of_nodes in edges_power_grid:
    power_grid.add_edge(pair_of_nodes[0],pair_of_nodes[1])
```

```python
C_power = nx.average_clustering(power_grid)
print(f"Coeficient de clustering mitjà: {C_power}")
L_power = nx.average_shortest_path_length(power_grid)
print(f"Longitud mitjana del camí més curt: {L_power}")
```

```python
# nodes de power grid random = nodes de power grid
power_grid_random = nx.Graph()
num_edges_power_grid_random = len(power_grid.edges())
power_grid_random.add_nodes_from(nodes_power_grid)
added_edges = 0
while added_edges < num_edges_power_grid_random:
    node_source = rnd.choice(nodes_power_grid) # Bibliografia: [56]
    node_target = rnd.choice(nodes_power_grid) # Bibliografia: [56]
    while node_source == node_target:
        node_target = rnd.choice(nodes_power_grid) # Bibliografia: [56]
    if not power_grid_random.has_edge(node_source, node_target):
        power_grid_random.add_edge(node_source, node_target)
        added_edges += 1
```

```python
C_power_random = nx.average_clustering(power_grid_random)
print(f"Coeficient de clustering mitjà: {C_power_random}")
```

```python
# Per calcular L cal un graf connex # Bibliografia: [22]
print(nx.is_connected(power_grid_random))# Bibliografia: [50]
```

```python
# Es buscara el subgraf connex més gran i es calcularà L
largest_cc = max(nx.connected_components(power_grid_random), key=len) # Bibliografia: [51]
largest_subgraph = power_grid_random.subgraph(largest_cc) # Bibliografia: [52]
print(nx.is_connected(largest_subgraph))
```

La imatge és una metèfora visual de la consideració de Kleinberg sobre una xarxa on els nodes s'enllaçen de manera aleatròria tenint en compte que la probabilitat d'enllaç disminueix amb la distància. Aquesta probabilitat d'enllaç entre nodes segueix una llei de potència amb un exponent paramètric <span>&#947;</span> = 2.

Quan aquest paràmetre es exactament 2, el graf/xarxa es troba en equilibri. És a dir, els _short-cuts_ antre dos nodes (a qualsevol distància) es poden trobar amb poca informació. Tot i que la probabilitat de connexió disminueix quan més lluny es va, també cal tenir en compte que com més lluny es va, més nodes hi ha per connectar-se.

Si <span>&#947;</span> > 2, la probabilitat de que dos nodes s'enllaçin disminueix molt ràpidament amb la distància, de tal manera que només els nodes propers entre si tenen alguna possibilitat de connectar-se.

Si <span>&#947;</span> < 2, el graf presenta el problema que presenta el model beta de Watts-Strogatz. Existeixen camins curts però no es poden trobar amb poca informació.

En observar la imatge, es pot veure, que aquesta representa la cauística de <span>&#947;</span> = 2. És a dir, la 9ª avinguda es el nostre punt de vista (el nostre node) i com més ens allunyem del nostre carrer més llocs diferents apareixen en el mapa (cada lloc diferent és un node) però com més lluny mirem des de la 9ª avinguda més es redueix la probabilitat de que ens connectem a aquell lloc on mirem. En altres paraules, És més fàcil (més probable) que si ens trobem a la 9ª avinguda, ens connectem amb la 10ª avinguda que amb un altre carrer d'un altre país.
